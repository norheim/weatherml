```julia
df = DataFrame(CSV.File("2352763.csv", header=true, types=Dict(:HourlyDryBulbTemperature=>Float64, :HourlyStationPressure=>Float64)));
```

```julia
dfh = filter(row -> !(strip(row[:REPORT_TYPE]) in ["SOD", "SOM"]), df) # these are the entries that correspond to daily and monthly data which we exclude;
```

```julia
pressure = Impute.interp(dfh[!, ["DATE","HourlyStationPressure"]]);
```

```julia
humidity = Impute.interp(dfh[!, ["DATE","HourlyRelativeHumidity"]]);
```

```julia
temp = Impute.interp(dfh[!, ["DATE","HourlyDryBulbTemperature"]]);
```

```julia
d0 = filter(row -> DateTime(2018, 1, 6, 0, 0)<=row[:DATE]<=DateTime(2018, 1, 15, 0, 0), pressure);
d1 = filter(row -> DateTime(2018, 1, 6, 0, 0)<=row[:DATE]<=DateTime(2018, 1, 15, 0, 0), temp);
d2 = filter(row -> DateTime(2018, 1, 6, 0, 0)<=row[:DATE]<=DateTime(2018, 1, 15, 0, 0), humidity);
```

```julia
l = @layout [a ; b ; c]
p1=plot(d0[!, "DATE"], d0[!, "HourlyStationPressure"])
p2=plot(d1[!, "DATE"], d1[!, "HourlyDryBulbTemperature"])
p3=plot(d2[!, "DATE"], d2[!, "HourlyRelativeHumidity"])
plot(p1, p2, p3, layout = l)
```
